## Задача: 
### Допустим, мы хотим собрать датафрейм, который будет включать в себя: первые три строки исходного датафрейма, строки 6-8 (включительно), строки 16-19 (включительно) и колонки: App, Size, Genres, Current Ver в соответствующем порядке.

### По сути, это будет выглядеть как объединение трех, состоящих из нескольких строк, датафреймов, являющимися срезами из исходных данных.

### В качестве ответа загрузи полученный датафрейм в формате csv. При сохранении финального датафрейма, колонка с индексами должна остаться.

```python
import pandas as pd
play_store = pd.read_csv("playstore_2.csv")
```

## Для начало я открыл таблицу playstore

```python
play_store
```

## Решение:

```python
#переменные ↓
pl_04=play_store.loc[0:4]
pl_69=play_store.loc[6:9]
pl_1620=play_store.loc[16:20]
colums=["App","Size","Genres","Current Ver"]
```

```python
# через concat объединим ↓
import pandas as pd
final_play_store=pd.concat([pl_04,pl_69,pl_1620],ignore_index=True)[colums]
```
